```python
import numpy as np
import GPy

import matplotlib.pyplot as plt
plt.ion()

%matplotlib inline

'''
From http://gpss.cc/gpss13/assets/lab1.pdf
'''
```

```python
k = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
k.plot()
```

```python
N = 3
T_final = 60
T_stop = 10 + np.random.choice([1,2,3,4,5], N, p=[0.1, 0.2, 0.2, 0.3, 0.2])

mu_0 = np.zeros(2)
S_0 = np.eye(2)

L = np.random.multivariate_normal(mu_0,S_0,N)
```

```python
T_stop
```

```python
mu_1 = 10.0
mu_2 = 0.1
# mu_3 = 10.0
mu_4 = 0.1

for i in range(N):
    X = np.arange(T_final) # 500 points evenly spaced over [0,1]

    ind_T_stop = (X >= T_stop[i]).astype(int)
    
    mu_3 = mu_1 - mu_2*T_stop[i]
    
    # vector of the means
    mu = (1-ind_T_stop) * (mu_1 - mu_2*X) \
         + ind_T_stop * mu_3*np.exp(-mu_4*(X - T_stop[i]))
    X_ = X[:, None]
    C = k.K(X_,X_) # covariance matrix

    # Generate 10 sample path with mean mu and covariance C
    Z = np.random.multivariate_normal(mu,C,20)

    plt.figure() # open new plotting window
    mean = Z.mean(axis=0)
    std = Z.std(axis=0)
    plt.plot(X,mean)
    plt.fill_between(X,mean-std,mean+std, alpha=0.1)
    plt.axvline(x=T_stop[i])
    plt.text(T_stop[i],0,'Stop')
    plt.title('Patient=%d'%i)
#     for j in range(20):
#         plt.plot(X[:],Z[j,:])
```
